```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sinplot()
```

```python
sns.set(palette="Set1")
sinplot()
```

```python
sns.set_style(style='whitegrid')
sinplot()
```

```python
sinplot()
sns.despine() #
```

# Seaborn figure styles
There are five preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks.  
figure

```python
sns.set_style('whitegrid')#darkgrid, whitegrid, dark, white, and ticks.
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
sns.set_style('white')
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="husl")
sns.despine(left=True)
```

# Temporarily setting figure style
use the axes_style() function in a with statement to temporarily set plot parameters

```python
with sns.axes_style("darkgrid"): 
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".8",'axes.linewidth': 1.5})
sinplot()
```
